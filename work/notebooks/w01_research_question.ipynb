## 1. My lane (or freestyle) and why

*Name your lane — or say 'freestyle' and describe your own question. One short paragraph: why this one?*

### CTR / Engagement Opportunity Scoring

I am choosing the CTR / Engagement Opportunity Scoring lane because it directly addresses a critical pain point for site owners: pages that successfully attract traffic (high sessions/impressions) but fail to keep usersengaged (low scroll/engagement rates). This traffic represents an underutilized asset that only needs refinement. While we cannot prove exactly what causes the low engagement without causal experiments, our model acts as a decision-support system to pinpoint which pages have the largest performance gaps relative to their position, helping editors prioritize where to investigate and apply improvements first.

## 2. The question: decision, action, cost of a wrong call

*What decision does your work improve? Who acts on it? What does a wrong recommendation cost?*

* **What decision does this improve?** -> Helping Content Editors to decide what page that need to be improve.
* **Who acts on the output, and what do they do?** Content Editors dan SEO Specialist. They search on the cause of the low engagement rate and try to improve it. 
* **What does a wrong recommendation cost?** Editors time is wasted reviewing and updating pages that are already performing optimally or have reached their natural maximum CTR for their current position..
* **Why does data or ML help at all?** Simple manual rules (like 'fix all pages with CTR < 1%') are inaccurate because CTR depends heavily on the average position of the page in the SERP (a page naturally ranked 1st has a much higher CTR compared to rank 9). ML helps model this non-linear relationship objectively based on historical data groups.

## 3. Quick look at the data (2-3 real numbers)

*Load the starter CSV below and show 2-3 real numbers that make your lane look worth the next 7 weeks.*

```python
import os
import pandas as pd

# Load data
data_paths = [
    "../../data/raw/content_refresh_anonymized.csv",
    "../data/raw/content_refresh_anonymized.csv",
    "data/raw/content_refresh_anonymized.csv"
]
df = None
for path in data_paths:
    if os.path.exists(path):
        df = pd.read_csv(path)
        print(f"Load success: {path}")
        break

if df is None:
    raise FileNotFoundError("File content_refresh_anonymized.csv not found.")

total_rows = len(df)
mean_ctr = df["ctr"].mean()
print(f"Total rows: {total_rows:,}")
print(f"CTR Average: {mean_ctr:.4f}%\n")

print("Average CTR & 90d impression based on position_tier:")
tier_stats = df.groupby("position_tier").agg(
    avg_ctr=("ctr", "mean"),
    avg_impressions=("impressions_90d", "mean"),
    total_pages=("content_id", "count")
)
print(tier_stats)
print()

top3 = df[df.position_tier == "top_3"]
page1 = df[df.position_tier == "page_1"]

top3_underperform = len(top3[(top3.impressions_90d >= 100) & (top3.clicks_90d == 0)])
page1_underperform = len(page1[(page1.impressions_90d >= 500) & (page1.ctr < 0.2)])

print(f"Top 3 with >= 100 impressions but the click is 0: {top3_underperform}")
print(f"Page 1 with >= 500 impressions but the CTR (< 0.2%): {page1_underperform}")
```

## 4. Careful words: what I can and can't claim

*Write what your work will be able to say (observed, directional, decision-support) — and what it never will (causal proof, 'predicting Google').*

## Safe Claims

- "Page A is observed to have a significantly lower CTR compared to the average of other pages within the same position tier."
- "We detect a structural relationship (correlation) between user engagement rate and certain content properties."
- "This model functions as a decision-support tool to prioritize which pages should be reviewed."

## Unsafe Claims

- "Changing this page’s meta description is guaranteed to improve its ranking position on Google."
- "This model successfully predicts the Google Search ranking algorithm."
- "Content change A is proven to causally increase click conversion by X%."
